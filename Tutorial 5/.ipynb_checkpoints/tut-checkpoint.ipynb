# 2. To Implement the Initial Version of Fourier Transform (FT) & Inverse Fourier Transform (IFT)

With the support for complex numbers, we can use the original definition of fourier transform to implement its corresponding function in python.

The 1-D continuous formula is as followed:

Continuous Fourier Transform (CFT_1d):

$$
F(\mu) = \mathcal{F}\{f(t)\} = \int^{\infty}_{-\infty} f(t) e^{-j 2 \pi \mu t} dt
$$

Continuous Inverse Fourier Transform (CIFT_1d):

$$
f(t) = \mathcal{F}^{-1}\{F(\mu)\} = \int^{\infty}_{-\infty} F(\mu) e^{j 2 \pi \mu t} d\mu
$$

- For the integration, we can utilize the `scipy.integrate` package for doing integration.
- Note that although the return value of the transformed function in frequency domain could be complex, we may expect its input is real.
- Using the fact that $\mathcal{F}^{-1}f = (\mathcal{F}f)^- = \mathcal{Ff^-}$, we can avoid re-defining a IFT but utilized the defined FT.
- For old version of `scipy.integrate`, it require the function to be of real input. You can call `cquad` where I separately calculate the real part and the imaginary part.
- The second part of quad's return is the error. We take result[0] for the actual result.

```python
from scipy.integrate import quad
from typing import Callable

def cquad(f: Callable[[float], complex], a: float, b:float):   # complex_quadrature
    def real_func(x):
        return np.real(f(x))
    def imag_func(x):
        return np.imag(f(x))
    real_integral = quad(real_func, a, b)
    imag_integral = quad(imag_func, a, b)
    return (real_integral[0] + 1j * imag_integral[0], (real_integral[1:], imag_integral[1:]))

my_quad = quad  # use `cquad` if your scipy version is too old

def CFT_1d(f: Callable[[float], complex])->Callable[[float],complex]:
    def result_func(mu):
        return my_quad(lambda t: f(t) * np.exp(- 1j * 2 * np.pi * mu * t), -np.inf, np.inf)[0]
    return result_func

def CIFT_1d(F: Callable[[float], complex])->Callable[[float], complex]:
    def result_func(t): # with conjugate complex exponential compared with fft
        return my_quad(lambda mu: F(mu) * np.exp(1j * 2 * np.pi * mu * t), -np.inf, np.inf)[0] # quite symmetric except that we exchange mu and t, F and f
    return result_func

# def CIFT_1d(F: Callable[[float], complex])->Callable[[float], complex]: # Another way of defining
#     F2 = CFT_1d(F)
#     def result_func(t):
#         return F2(-t)
#     return result_func
```

We define a simple "rectangle" function here and check what's the function transformed to the frequency domain looks like. Here we use `matplotlib` to plot the shape of the function.

```python
import matplotlib.pyplot as plt

def rect(t: float):
    if t >= - .5 and t <= .5:
        return 1
    else:
        return 0

rect_vec = np.vectorize(rect)

x = np.linspace(-1, 1, num=500)
plt.plot(x, rect_vec(x), label=r'$\pi(t)$')
plt.legend()
plt.show()
```

Here we plot the shape of $F(\mu)$. You shall see that `matplotlib` shows an warning that "Casting complex values to real discards the imaginary part
  return np.asarray(x, float)", since it cannot plot the imaginary part on a x-y plane. However, it's okay since you can proved that the imaginary part is 0.

```python
frect = CFT_1d(rect)

frect_vec = np.vectorize(frect)

x = np.linspace(-5, 5, num=500)
y = frect_vec(x)
plt.plot(x, y, label=r'$\mathcal{F}\{\pi(t)\}$')
plt.legend()
plt.show()
```

Let's check its spectrum (just taking the absolution value as there isn't any imaginary part).

```python
plt.plot(x, np.abs(y), label=r'$|\mathcal{F}\{\pi(t)\}|$')
plt.legend()
plt.show()
```

Now we can check whether the function is the same if we take the inverse fourier transform:
You can see that the speed is so slow and the result is even **incorrect**. This is because that after two rounds of integration, the precision is not enough and the integral diverges. Hence it shows the necessity to introduce a reliable fast numerical method for calculating the `fft` and `ifft` (Note that at this step you need to wait 10 to 20 seconds for the results. You can see that it's really slow).

```python
rect_back = CIFT_1d(frect)
a = rect(0)
b = rect_back(0)
print(a)
print(b)
print(a==b)
```

Actually for our input x, it's a linear space that is discretized, so it's okay to use the **Discrete Version of Fourier Transform**. Finite terms of summation would be a lot more faster. Here we can utilize "Fast Fourier Transform" and "Inverse Fast Fourier Transform", which could speed up the calculation. These are implemented by `scipy.fft.fft` and `scipy.fft.ifft` (in the old version of SciPy, you can find parts of the contents of `scipy.fft` in `scipy.fftpack`, which is now a legacy module).

```python
from scipy.fft import fft, ifft

t = np.linspace(-1, 1, num=500)
ft = rect_vec(t)
Fmu = fft(ft)
ft_back = ifft(Fmu)

plt.plot(t, ft, label=r'$\pi(t)$')
plt.legend()
plt.show()

plt.plot(t, ft_back, label=r'$|\mathcal{F}^{-1}\{\mathcal{F}\{\pi(t)\}\}|$')
plt.legend()
plt.show()
```

We use the FFT provided by SciPy just for the convenience of extending to arbitrary n-dimension (which would be discussed later). Numpy also provides a very quick version of Discrete Fourier Transform (DFT) and its inverse one (IDFT), which utilizes the algorithm of Fast Fourier Transform (FFT), and you might be curious about its trick of implementation. We have a very nice [tutorial notebook](understanding_dft_n_fft.ipynb) for understanding the original version of DFT by [Jake VanderPlas](http://vanderplas.com/), so I will not type the formula and its implementation. You may check this jupyter notebook for better understanding DFT and FFT. Below is a copy of `FFT_vectorized` he implemented, and we can use it to try again:

```python
def FFT_vectorized(x):
    """A vectorized, non-recursive version of the Cooley-Tukey FFT"""
    x = np.asarray(x, dtype=float)
    N = x.shape[0]

    if np.log2(N) % 1 > 0:
        raise ValueError("size of x must be a power of 2")

    # N_min here is equivalent to the stopping condition above,
    # and should be a power of 2
    N_min = min(N, 32)
    
    # Perform an O[N^2] DFT on all length-N_min sub-problems at once
    n = np.arange(N_min)
    k = n[:, None]
    M = np.exp(-2j * np.pi * n * k / N_min)
    X = np.dot(M, x.reshape((N_min, -1)))

    # build-up each level of the recursive calculation all at once
    while X.shape[0] < N:
        X_even = X[:, :X.shape[1] // 2]
        X_odd = X[:, X.shape[1] // 2:]
        factor = np.exp(-1j * np.pi * np.arange(X.shape[0]) / X.shape[0])[:, None]
        X = np.vstack([X_even + factor * X_odd,
                       X_even - factor * X_odd])

    return X.ravel()

t = np.linspace(-1, 1, num=512) # make sure the number is a power of 2
ft = rect_vec(t)
Fmu = FFT_vectorized(ft)
ft_back = 1 / len(t) * FFT_vectorized(Fmu) # the original function is even: F^-1(Ff) = (FFf)^- = FFf, with the normalization coefficient 1/M

plt.plot(t, ft, label=r'$\pi(t)$')
plt.legend()
plt.show()

plt.plot(t, ft_back, label=r'$|\mathcal{F}^{-1}\{\mathcal{F}\{\pi(t)\}\}|$')
plt.legend()
plt.show()
```

# 3. To Extend the 1-D FFT & IFFT to 2-D or n-D ones

In `scipy.fft`, the functions `fft2` and `ifft2` provide 2-D FFT and IFFT, respectively. Similarly, `fftn` and `ifftn` provide N-D FFT, and IFFT, respectively. It's useful that this function can automatically recognize the input dimensional information.

For real-input signals, we have `rfft` for the 1-D case. We also have the functions `rfft2` and `irfft2` for 2-D real transforms; `rfftn` and `irfftn` for N-D real transforms.

The example below demonstrates a 2-D IFFT and plots the resulting (2-D) time-domain signals.

```python
from scipy.fft import ifftn
import matplotlib.cm as cm
N = 30
f, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3, sharex='col', sharey='row', figsize=(15, 10))

xf = np.zeros((N,N))
xf[0, 5] = 1
xf[0, N-5] = 1
Z = ifftn(xf)

ax1.imshow(xf, cmap=cm.gray)
ax4.imshow(np.real(Z), cmap=cm.gray)

xf = np.zeros((N, N))
xf[5, 0] = 1
xf[N-5, 0] = 1
Z = ifftn(xf)

ax2.imshow(xf, cmap=cm.gray)
ax5.imshow(np.real(Z), cmap=cm.gray)

xf = np.zeros((N, N))
xf[5, 10] = 1
xf[N-5, N-10] = 1
Z = ifftn(xf)

ax3.imshow(xf, cmap=cm.gray)
ax6.imshow(np.real(Z), cmap=cm.gray)

plt.show()
```

# 4. To Use `fftn` & `ifftn` for Digital Image Processing

With the help of `scipy.fft.fftn` and `scipy.fft.ifftn`, we can now convert an image to its frequency domain. Note that you can also utilize `numpy.fft.fft2` and `numpy.fft.ifft2` if yo're just handling the 2-D image. We also need to shift the image at its frequency domain with `fftshift`. We can see that the result in the middle is very similar to that of the lecture slides. However, as the source image is just screen-shot rather than the original image, there is still some difference.

```python
# -*- coding: utf-8 -*-
from scipy.fft import fftn, fftshift, ifftshift # It's also okay to use `numpy.fft`
import cv2 as cv
from matplotlib import pyplot as plt

def forward_backward(img_path: str):
    # read the image
    img = cv.imread(img_path, 0)

    # fourier transform
    Fmu = fftn(img)
    fshift = fftshift(Fmu)

    res = np.log(np.abs(fshift))    # the result is just for plotting -- here we get the spectrum and depolarized it

    # inverse fourier transform
    ishift = ifftshift(fshift)
    iimg = ifftn(ishift)
    iimg = np.real(iimg)                # keep the real part (when there is no imaginary part, we can also use `iimg.astype(float)`)
    assert np.all(np.imag(iimg) == 0)   # this could pass the check as the imaginary part equals 0 (since the original function output is real)

    return img, res, iimg

# plot the result
def plot(vals):
    n = len(vals)
    h = 6 * n
    plt.figure(figsize=(18,h))
    base = n * 100 + 30
    for i, (img, res, iimg) in enumerate(vals):
        plt.subplot(base + i*3+1), plt.imshow(img, 'gray'), plt.title('(a) original')
        plt.subplot(base + i*3+2), plt.imshow(res, 'gray'), plt.title('(b) FFT Spectrum (shift)')
        plt.subplot(base + i*3+3), plt.imshow(iimg, 'gray'), plt.title('(c) IFFT (shift back)')
    plt.show()

imgs = ['rectangle.png', 'woman.png']

plot([forward_backward(img) for img in imgs])
```

Note that if we don't do the shifting, the result is not centralized. The `fftshift` rearranges the outputs of fft, fft2, and fftn by moving the zero-frequency component to the center of the array. It is useful for visualizing a Fourier transform with the zero-frequency component in the middle of the spectrum. The `ifftshift` do the inverse process.

```python
def forward_backward_no_shift(img_path: str):
    # read the image
    img = cv.imread(img_path, 0)

    # fourier transform
    Fmu = fftn(img)

    res = np.log(np.abs(Fmu))    #  # the result is just for plotting -- here we get the spectrum and depolarized it

    # inverse fourier transform
    iimg = ifftn(Fmu)
    iimg = np.real(iimg)

    return img, res, iimg



# plot the result
plot([forward_backward_no_shift(img) for img in imgs])
```

Also note that we take the `log` function here over the transformed result just to depolarize it, so that it could be shown clearly. You can see that there is a "white dot" in the middle.

```python
def forward_backward_no_depolarize(img_path: str):
    # read the image
    img = cv.imread(img_path, 0)

    # fourier transform
    Fmu = fftn(img)
    fshift = fftshift(Fmu)

    res = np.abs(np.abs(fshift))    # didn't depolarize it -> hard to see the result in frequency domain

    # inverse fourier transform
    ishift = ifftshift(fshift)
    iimg = ifftn(ishift)
    iimg = np.real(iimg) # keep the real part

    return img, res, iimg



# plot the result
plot([forward_backward_no_depolarize(img) for img in imgs])
```

# 5. To Consider the Influence of Spectrum & Phase

Using the following formula provided in the lecture notes, we can separate the function in the frequency domain into the **spectrum** part and the **phase** part. Note that the symbol for the frequency are $\mu$ and $\nu$, but for convenience for notation we change them to $u$ and $v$ as they are included in *ASCII*:

2-D DFT in polar form:
$$
F(u, v) = |F(u,v)| e ^ {j \phi(u,v)} = C(u,v) e ^ {j \phi(u,v)}
$$

Fourier Spectrum:
$$
C(u,v) = |F(u,v)| = (R^2(u,v) + I^2(u,v))^{1/2}
$$

Phase Angle:
$$
\phi(u,v) = tan^{-1}(I(u,v) / R(u,v))
$$

Note that here I provide a faster version of getting the spectrum and the phase using the built-in function rather than strictly sticking to these formula. You can switch the function as you wish.

```python
from typing import Union, Tuple

Complex = Union[np.ndarray, complex]
Float = Union[np.ndarray, float]

def get_c_and_phi_standard(x: Complex)->Tuple[Float, Float]:    # i.e. to_polar_form
    i = np.imag(x)
    r = np.real(x)
    return np.sqrt(r**2 + i**2), np.arctan2(i, r)   # it takes two variables compared with `np.arctan`, which only takes i/r and could have some sign problem
                                                    # if you use `np.arctan(i/r), you shall encounter the sign issue`

def get_c_and_phi_quick(x: Complex)->Tuple[Float, Float]:    # a faster version as it use the built-in function
    return np.abs(x), np.angle(x)

get_c_and_phi = get_c_and_phi_quick    # You may try `get_c_and_phi_quick` or `get_c_and_phi_standard`

def from_c_and_phi(c: Float, phi: Float)->Complex:      # i.e. from_polar_form
    return c * np.exp(1j * phi)

def use_c_only(x: Complex)->Complex:
    c, _ = get_c_and_phi(x)                 
    return from_c_and_phi(c, 0)         # set phi as a constant: 0

def use_phi_only(x: Complex)->Complex:
    _, phi = get_c_and_phi(x)
    return from_c_and_phi(1, phi)       # set c as a constant: 1


# read the image
def plot_self_recover(img_path: str):
    img = cv.imread(img_path, 0)

    # fourier transform
    Fmu = fftn(img)
    fshift = fftshift(Fmu)

    fc = use_c_only(fshift)
    fphi = use_phi_only(fshift)

    res_fc = np.log(np.abs(fc))       # equals the spectrum, i.e. np.abs(fshift). It will automatically calculate the spectrum
    res_fphi = np.log(np.abs(fphi))

        # inverse fourier transform
    ishiftc = ifftshift(fc)
    iimgc = ifftn(ishiftc)
    iimgc = np.real(iimgc)

    ishiftphi = ifftshift(fphi)
    iimgphi = ifftn(ishiftphi)
    iimgphi = np.real(iimgphi)

    plt.figure(figsize=(18, 12))
    plt.subplot(231), plt.imshow(img, 'gray'), plt.title('(a) original')
    plt.subplot(232), plt.imshow(res_fc, 'gray'), plt.title('(b) FFT Spectrum (shift)')
    plt.subplot(233), plt.imshow(iimgc, 'gray'), plt.title('(c) IFFT Spectrum (shift)')
    plt.subplot(235), plt.imshow(res_fphi, 'gray'), plt.title('(d) FFT Phase (shift)')
    plt.subplot(236), plt.imshow(iimgphi, 'gray'), plt.title('(e) IFFT Phase (shift back)')
    plt.show()

for img in imgs:
    plot_self_recover(img)
```

Now we try to combine the spectrum of one image and the phase of the other image, and see what will happen:

```python
img0 = cv.imread(imgs[0], 0)
img1 = cv.imread(imgs[1], 0)

# fourier transform
Fmu0 = fftn(img0)
fshift0 = fftshift(Fmu0)

Fmu1 = fftn(img1)
fshift1 = fftshift(Fmu1)

c0, phi0 = get_c_and_phi(fshift0)
c1, phi1 = get_c_and_phi(fshift1)

c0phi1 = from_c_and_phi(c0, phi1)
c1phi0 = from_c_and_phi(c1, phi0)

# inverse fourier transform
ishift01 = ifftshift(c0phi1)
iimg01 = ifftn(ishift01)
iimg01 = np.real(iimg01)

ishift10 = ifftshift(c1phi0)
iimg10 = ifftn(ishift10)
iimg10 = np.real(iimg10)

plt.figure(figsize=(18, 18))
plt.subplot(221), plt.imshow(img0, 'gray'), plt.title('(a) Image 0')
plt.subplot(222), plt.imshow(img1, 'gray'), plt.title('(b) Image 1')
plt.subplot(223), plt.imshow(iimg01, 'gray'), plt.title('(c) Spectrum 0 + Phase 1')
plt.subplot(224), plt.imshow(iimg10, 'gray'), plt.title('(d) Spectrum 1 + Phase 0')
plt.show()
```

We can also check the translation and rotation property using the spectrum and the phase information. Here we use the rectangle image as an example.

```python
def origin_spectrum_phase(img_path: str):
    # read the image
    img = cv.imread(img_path, 0)

    # fourier transform
    Fmu = fftn(img)
    fshift = fftshift(Fmu)

    fc = use_c_only(fshift)
    fphi = use_phi_only(fshift)

    spec = np.log(np.abs(fc))       # equals the spectrum, i.e. np.abs(fshift). It will automatically calculate the spectrum
    phase = np.log(np.abs(fphi))

    return img, spec, phase

# plot the result
def plot(vals):
    n = len(vals)
    h = 6 * n
    plt.figure(figsize=(18,h))
    base = n * 100 + 30
    for i, (img, spec, phase) in enumerate(vals):
        plt.subplot(base + i*3+1), plt.imshow(img, 'gray'), plt.title('(a) original')
        plt.subplot(base + i*3+2), plt.imshow(spec, 'gray'), plt.title('(b) FFT Spectrum (shift)')
        plt.subplot(base + i*3+3), plt.imshow(phase, 'gray'), plt.title('(c) FFT Phase (shift)')
    plt.show()

imgs = ['rectangle.png', 'rectangle_translate.png', 'rectangle_rotate2.png']

plot([origin_spectrum_phase(img) for img in imgs])
```

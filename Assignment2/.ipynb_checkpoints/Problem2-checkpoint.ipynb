```python
import cv2
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

def estimateWhiteNoise(img):
    # 如果图像是彩色的，将其转换为灰度图
    if len(img.shape) == 3:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # 计算图像的梯度
    sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
    sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
    gradient_magnitude = np.sqrt(sobelx**2 + sobely**2)
    
    # 估计噪声标准差
    noise_std_dev = np.std(gradient_magnitude)
    
    return noise_std_dev

# 加载退化图像
img1 = cv2.imread('pro2_satellite_1_degraded.tiff', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('pro2_satellite_2_degraded.tiff', cv2.IMREAD_GRAYSCALE)

# 估计噪声标准差
noise_std_dev1 = estimateWhiteNoise(img1)
noise_std_dev2 = estimateWhiteNoise(img2)

print(f'Estimated noise standard deviation for image 1: {noise_std_dev1}')
print(f'Estimated noise standard deviation for image 2: {noise_std_dev2}')
```

```python
def estimateGaussianImpulse(img):
    # 如果图像是彩色的，将其转换为灰度图
    if len(img.shape) == 3:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # 计算图像的离散傅里叶变换（DFT）
    dft = np.fft.fft2(img)
    dft_shift = np.fft.fftshift(dft)
    magnitude_spectrum = np.abs(dft_shift)
    
    # 获取幅值谱的中心部分
    rows, cols = magnitude_spectrum.shape
    crow, ccol = rows // 2 , cols // 2
    center_region = magnitude_spectrum[crow - 30:crow + 30, ccol - 30:ccol + 30]
    
    # 计算中心部分的二阶矩
    x = np.linspace(-15, 15, 30*2)
    y = np.linspace(-15, 15, 30*2)
    X, Y = np.meshgrid(x, y)
    
    # 计算二阶矩
    sigma_x = np.sqrt(np.sum(center_region * (X**2)) / np.sum(center_region))
    sigma_y = np.sqrt(np.sum(center_region * (Y**2)) / np.sum(center_region))
    
    # 取平均作为估计的标准差
    sigma = (sigma_x + sigma_y) / 2
    
    return sigma

# 加载退化图像
img1 = cv2.imread('pro2_satellite_1_degraded.tiff', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('pro2_satellite_2_degraded.tiff', cv2.IMREAD_GRAYSCALE)

# 估计高斯脉冲响应的标准差
gaussian_std_dev1 = estimateGaussianImpulse(img1)
gaussian_std_dev2 = estimateGaussianImpulse(img2)

print(f'Estimated Gaussian standard deviation for image 1: {gaussian_std_dev1}')
print(f'Estimated Gaussian standard deviation for image 2: {gaussian_std_dev2}')
```

```python
def wienerFilter(img, gaussian_std_dev, noise_std_dev, k_values):
    # 如果图像是彩色的，将其转换为灰度图
    if len(img.shape) == 3:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # 计算图像的离散傅里叶变换（DFT）
    dft = np.fft.fft2(img)
    dft_shift = np.fft.fftshift(dft)
    
    # 创建高斯脉冲响应的频域表示
    rows, cols = img.shape
    crow, ccol = rows // 2 , cols // 2
    x = np.linspace(-ccol, ccol, cols)
    y = np.linspace(-crow, crow, rows)
    X, Y = np.meshgrid(x, y)
    gaussian_filter = np.exp(-((X**2 + Y**2) / (2 * gaussian_std_dev**2)))
    
    # 初始化结果列表
    results = []
    
    # 执行维纳滤波
    for k in k_values:
        H_conj = np.conj(gaussian_filter)
        SNR = (gaussian_filter**2) / noise_std_dev**2
        wiener_filter = H_conj / (gaussian_filter**2 + k / SNR)
        dft_wiener = dft_shift * wiener_filter
        dft_wiener_shift = np.fft.ifftshift(dft_wiener)
        img_wiener = np.fft.ifft2(dft_wiener_shift)
        img_wiener = np.abs(img_wiener)
        results.append((k, img_wiener))
    
    return results

# 加载退化图像
img1 = cv2.imread('pro2_satellite_1_degraded.tiff', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('pro2_satellite_2_degraded.tiff', cv2.IMREAD_GRAYSCALE)

# 执行维纳滤波
k_values = [0.01, 0.1, 1]
results1 = wienerFilter(img1, gaussian_std_dev1, noise_std_dev1, k_values)
results2 = wienerFilter(img2, gaussian_std_dev2, noise_std_dev2, k_values)

# 显示维纳滤波后的图像
plt.figure(figsize=(15, 10))

for i, (k, img_wiener) in enumerate(results1):
    plt.subplot(2, 3, i + 1)
    plt.title(f'Image 1, K={k}')
    plt.imshow(img_wiener, cmap='gray')
    plt.axis('off')

for i, (k, img_wiener) in enumerate(results2):
    plt.subplot(2, 3, i + 4)
    plt.title(f'Image 2, K={k}')
    plt.imshow(img_wiener, cmap='gray')
    plt.axis('off')

plt.show()
```

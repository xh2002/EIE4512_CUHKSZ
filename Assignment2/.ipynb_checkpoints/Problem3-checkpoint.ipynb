```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import median_filter

def apply_median_filter(img_path, N):
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    filtered_img = median_filter(img, size=N)
    return img, filtered_img

img1_path = 'pro3_radiograph_1.jpg'
img2_path = 'pro3_radiograph_2.jpg'

N = 5  # Example window size, adjust based on results
img1, filtered_img1 = apply_median_filter(img1_path, N)
img2, filtered_img2 = apply_median_filter(img2_path, N)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(filtered_img1, cmap='gray')
plt.title('Filtered Image 1')
plt.subplot(1, 2, 2)
plt.imshow(filtered_img2, cmap='gray')
plt.title('Filtered Image 2')
plt.show()

# Report N
print(f'Chosen window size N: {N}')
```

```python
def compute_dft(img):
    dft = np.fft.fftshift(np.fft.fft2(img))
    magnitude_spectrum = np.log(np.abs(dft) + 1)
    return magnitude_spectrum

dft_magnitude1 = compute_dft(filtered_img1)
dft_magnitude2 = compute_dft(filtered_img2)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(dft_magnitude1, cmap='gray')
plt.title('DFT Magnitude Image 1')
plt.subplot(1, 2, 2)
plt.imshow(dft_magnitude2, cmap='gray')
plt.title('DFT Magnitude Image 2')
plt.show()
```

```python
def notch_filter(img, notch_centers, notch_width):
    rows, cols = img.shape
    notch_filter = np.ones((rows, cols))

    for center in notch_centers:
        row, col = center
        notch_filter[row-notch_width:row+notch_width, col-notch_width:col+notch_width] = 0
        notch_filter[rows-row-notch_width:rows-row+notch_width, cols-col-notch_width:cols-col+notch_width] = 0
    
    dft = np.fft.fftshift(np.fft.fft2(img))
    filtered_dft = dft * notch_filter
    filtered_img = np.fft.ifft2(np.fft.ifftshift(filtered_dft))
    filtered_img = np.abs(filtered_img)
    return notch_filter, filtered_dft, filtered_img

notch_centers = [(30, 30), (60, 60)]  # Example values, adjust based on DFT magnitude
notch_width = 5

notch_filter1, filtered_dft1, filtered_img1 = notch_filter(filtered_img1, notch_centers, notch_width)
notch_filter2, filtered_dft2, filtered_img2 = notch_filter(filtered_img2, notch_centers, notch_width)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(np.log(np.abs(filtered_dft1) + 1), cmap='gray')
plt.title('Filtered DFT Magnitude Image 1')
plt.subplot(1, 2, 2)
plt.imshow(np.log(np.abs(filtered_dft2) + 1), cmap='gray')
plt.title('Filtered DFT Magnitude Image 2')
plt.show()

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(filtered_img1, cmap='gray')
plt.title('Notch Filtered Image 1')
plt.subplot(1, 2, 2)
plt.imshow(filtered_img2, cmap='gray')
plt.title('Notch Filtered Image 2')
plt.show()
```

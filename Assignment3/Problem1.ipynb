```python
# Problem 1
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 
img1 = cv2.imread('./figs/pro1_page1.jpg')
img2 = cv2.imread('./figs/pro1_page2.jpg')
img3 = cv2.imread('./figs/pro1_page3.png')

# 
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.title('pro1_page1.jpg')
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))

plt.subplot(1, 3, 2)
plt.title('pro1_page2.jpg')
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))

plt.subplot(1, 3, 3)
plt.title('pro1_page3.png')
plt.imshow(cv2.cvtColor(img3, cv2.COLOR_BGR2RGB))

plt.show()
```

```python
def globalThresh(img):
    #  Otsu 
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    ret, binary_img = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    
    # 
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.title('Global Thresholding')
    plt.imshow(binary_img, cmap='gray')
    
    # 
    plt.subplot(1, 2, 2)
    plt.hist(gray.ravel(), bins=256)
    plt.axvline(x=ret, color='r', linestyle='dashed', linewidth=2)
    plt.title('Histogram')
    plt.show()
    
    return binary_img

```

```python
def localAdaptThresh(img):
    # 
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    binary_img = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                       cv2.THRESH_BINARY, 11, 2)
    
    # 
    plt.figure(figsize=(5, 5))
    plt.title('Local Adaptive Thresholding')
    plt.imshow(binary_img, cmap='gray')
    plt.show()
    
    return binary_img
```

```python
# 
print("Global Thresholding for pro1_page1.jpg and pro1_page2.jpg")
globalThresh(img1)
globalThresh(img2)

# 
```

```python
# 
print("Local Adaptive Thresholding for pro1_page1.jpg and pro1_page2.jpg")
localAdaptThresh(img1)
localAdaptThresh(img2)

# 
```

```python
# 
print("Local Adaptive Thresholding for pro1_page3.png")
localAdaptThresh(img3)
```
